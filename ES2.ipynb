```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("D:\\Attendance\\Attendance\\DSA\\attendance.csv")

# Calculate the mean attendance percentage for each student
df['Attendance'] = df[['02-04-2024', '03-04-2024','06-04-2024', '08-04-2024', '09-04-2024']].mean(axis=1)

# Plot a pie chart of attendance distribution
plt.figure(figsize=(8, 8))
plt.pie(df['Attendance'], labels=df['Name'], autopct='%1.1f%%', startangle=140)
plt.title('Overall Attendance Distribution')

# Adjust layout to increase spacing between title and plot
plt.tight_layout(pad=4.0)  # You can adjust the value of pad as needed

plt.axis('equal')
plt.show()
```

```python
# Filter the DataFrame for the student "Vinay Saunshi"
vinay_df = df[df['Name'] == 'Abhay Chougala']

# Calculate the mean attendance percentage for Vinay Saunshi
vinay_attendance = vinay_df[['02-04-2024', '03-04-2024','06-04-2024', '08-04-2024', '09-04-2024']].mean(axis=1).iloc[0]

# Plot a pie chart for Vinay Saunshi's attendance percentage
plt.figure(figsize=(6, 6))
plt.pie([vinay_attendance, 100 - vinay_attendance], labels=['Attendance', 'Absent'], autopct='%1.1f%%', startangle=140)
plt.title('Vinay Saunshi Attendance Percentage')
plt.axis('equal')
plt.show()
```
